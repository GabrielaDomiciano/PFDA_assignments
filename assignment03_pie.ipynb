```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

```python
## Let's take a look at the Dataset.
df = pd.read_csv('https://drive.google.com/uc?id=1AWPf-pJodJKeHsARQK_RHiNsE8fjPCVK&export=download')
df
```

```python
#The code takes the 'Email' column from the DataFrame df and assigns it to a new variable email_df. Then, display(email_df) shows the content of that column.
email_df = df['Email']
display(email_df)
```

```python
# The code extracts the domain from each email, counts how often each domain appears, and then displays a pie chart that shows the distribution of email domains in the dataset

# Function to extract the domain from emails
def extract_domain(email_list):
    domains = []
    for email in email_list:
        domain = email.split('@')[-1]  # Extract the part after the '@'
        domains.append(domain)
    return domains

# Extract domains from emails
email_domains = extract_domain(email_df)

# Count the frequency of each domain
domain_counts = Counter(email_domains)

# Create the pie chart
# Labels contains the unique domains (like gmail.com, yahoo.com).
# Sizes contains the count of how many times each domain appears.
labels = domain_counts.keys()
sizes = domain_counts.values()

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90) # autopct='%1.1f%%' adds the percentage of each domain to the chart.

plt.axis('equal')  # Creating the graph Matching the proportion to ensure the graph is circular

# show
plt.title('Distribution of Email Domains')
plt.legend(title='Email Domains ', labels=labels)
plt.savefig('/Users/gabrieladomicianoavellar/Desktop/PFDA_assignments/plot/distrib_email_domain.png')
plt.show()

```
