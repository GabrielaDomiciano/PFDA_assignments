```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import matplotlib.dates as mdates
```

```python
# Let's take a look at the Dataset.
weather_df = pd.read_csv('weatherreadings1.csv')
display(weather_df)
```

```python
temperature = weather_df[['reportStartDateTime', 'dryBulbTemperature_Celsius']]
# Show.
print(temperature)
```

```python
#formatting the data 
datetime = temperature[['reportStartDateTime']]
print(datetime)
```

```python
temperature_celsius = temperature[['dryBulbTemperature_Celsius']]
# Show.
print(temperature_celsius)
```

```python
# Using reportStartDateTime on the x-axis and dryBulbTemperature_Celsius on the y-axis. 
#The plot uses dashed lines (linestyle='dashed') and sets the line width to 2 (linewidth='2').
temperature.plot('reportStartDateTime' , 'dryBulbTemperature_Celsius', linestyle = 'dashed', linewidth = '2')

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

# Title.
plt.title('Temperature Over Time (℃)', fontdict = font1, loc = 'left')

# Axis labels.
plt.xlabel('Report Start Date Time', fontdict = font2, color = 'r')
plt.ylabel('Dry Bulb Temperature (℃)', fontdict = font2, color = 'r')
#rotated the x-axis labels by 90 degrees to make them easier to read if they are long.
plt.xticks(rotation=90)
plt.savefig('temperature_over_time.png')

```
